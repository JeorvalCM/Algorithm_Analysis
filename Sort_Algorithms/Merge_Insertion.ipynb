```python
def insertion_sort(array):
    """
    Docstring: function for insertion sort
    Input: array to be sorted
    Output: sorted array
    """
    for j in range (1,len(array)):
        key = array[j]
        i = j
        while i > 0 and array[i-1] > key:
            array[i] = array[i-1]
            i -= 1
        array[i] = key
        
def merge_sort(array):
    """
    Docstring: Function for the merge sort
    Input: an array to sort
    Output: a sorted array
    """
    #checks if the size of the elements greater than 1
    if len(array) >1: 
        #getting the mid point of the array
        mid = int(len(array)/2) 
        #declaring the left and right array
        Left = []
        Right = []
        #dividing for conquerer
        for i in range(0,mid):
            Left.append(array[i])
        for i in range(mid, len(array)):
            Right.append(array[i])
        #dividing recursively, first half
        merge_sort(Left) 
        #right half
        merge_sort(Right)
        #conquering, basycally getting all the subarrays together
        merge(array, Left,Right)
          
def merge(array, L, R):
    """
    Docstring: function of the merge sort to get the subarrays together
    input: arr: original array, L: left half, R: right half
    """
    i = 0
    j = 0
    z = 0
    # sorting, where the original array get the values from the temp
    while i < len(L) and j < len(R): 
        if L[i] < R[j]: 
            array[z] = L[i] 
            i+=1
        else: 
            array[z] = R[j] 
            j+=1
        z+=1

    # last checking, in case there is any element left
    while i < len(L): 
        array[z] = L[i] 
        i+=1
        z+=1

    while j < len(R): 
        array[z] = R[j] 
        j+=1
        z+=1
            
            
```

## 2. Compare the runtimes of each algorithm for several sizes of input. Populate the inputs with random numbers. For each size, make different simulations and compute the mean of the running time. Make a table with your results (Size, Insertion Sort Running Time, Merge Sort Running Time).

```python
#script for the test

mean_times = []
mean_time_merge = []
#list of the size from 100 to 15900
sizes = [i*100 for i in range(1,160)]
#loop to test in this size
for y in (sizes):
    tiempos = []
    tiempos_merge = []
    #loop to make an average of the same size of N
    for z in range(10):
        #making a random list
        x = [np.random.randint(1000) for x in range(y)]
        x2 = x[:]
        #getting the times for the insertion sort
        time1 = time()
        insertion_sort(x)
        time2 = time()-time1
        tiempos.append(time2)
        #getting the times for the merge sort
        time3 = time()
        merge_sort(x2)
        time4 = time()-time3
        tiempos_merge.append(time4)
    print(y)
    #getting mean times
    mean_times.append(np.mean(tiempos))
    mean_time_merge.append(np.mean(tiempos_merge))
    
```

```python
#making a dictionary to then transform it to a DataFrame
dataframe = {
    'Size': sizes,
    'Running_time_insertion': mean_times
    'Running_time_merge': mean_time_merge
}
```

```python
pruebas = pd.DataFrame(dataframe)
pruebas.to_csv('Pruebas_finales.csv')
```
